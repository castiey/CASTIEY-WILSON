## Challenge 1: Skewness

### Definition

Skewness is the third standardized central moment of a distribution, defined as:

$$\text{Skewness} = \gamma_1 = E\left[\left(\frac{X - \mu}{\sigma}\right)^3\right] = \frac{\mu_3}{\sigma^3}$$

where $\mu$ is the mean, $\sigma$ is the standard deviation, and $\mu_3$ is the third central moment. A symmetric distribution has $\gamma_1 = 0$; negative values indicate a left-skewed (long left tail) distribution, and positive values indicate a right-skewed distribution.

### Description

Skewness measures the degree of asymmetry in a return distribution. In financial time series, returns commonly exhibit **negative skewness**, meaning extreme negative returns (crashes) occur more frequently and with greater magnitude than extreme positive returns — a phenomenon at odds with the normality assumption embedded in many standard financial models.

### Diagnosis

Skewness can be detected through:
1. **Direct computation** of the Fisher-Pearson skewness coefficient — values significantly different from zero indicate asymmetry.
2. **Jarque-Bera (JB) Test** — a formal statistical test for normality that penalizes both skewness and excess kurtosis:

$$JB = \frac{n}{6}\left(S^2 + \frac{(K-3)^2}{4}\right)$$

where $S$ is skewness, $K$ is kurtosis, and $n$ is the sample size. Under the null hypothesis of normality, $JB \sim \chi^2(2)$. A p-value below 0.05 rejects normality.

```python
# ── DEMONSTRATION: Skewness ──────────────────────────────────────────────────

returns = data['Returns']

# Descriptive statistics
skew_val  = returns.skew()
kurt_val  = returns.kurtosis()          # excess kurtosis (normal = 0)
mean_val  = returns.mean()
std_val   = returns.std()

# Jarque-Bera formal test
jb_stat, jb_pvalue = stats.jarque_bera(returns)

print("=" * 50)
print("   SKEWNESS DIAGNOSTIC REPORT — AAPL Returns")
print("=" * 50)
print(f"  Mean Return        : {mean_val:.6f}")
print(f"  Std Deviation      : {std_val:.6f}")
print(f"  Skewness (γ₁)      : {skew_val:.4f}")
print(f"  Excess Kurtosis    : {kurt_val:.4f}")
print("-" * 50)
print(f"  Jarque-Bera Stat   : {jb_stat:.4f}")
print(f"  JB p-value         : {jb_pvalue:.2e}")
print("-" * 50)
if jb_pvalue < 0.05:
    print("  RESULT: Normality REJECTED at 5% significance.")
    print("  Returns are NOT normally distributed.")
print("=" * 50)
```

### Interpretation of Results

The computed skewness of AAPL log returns is **negative**, confirming a left-skewed distribution. The Jarque-Bera test rejects normality at any conventional significance level (p-value ≪ 0.05). The Q-Q plot reinforces this finding: the tails deviate sharply from the 45-degree reference line, particularly on the left (downside), indicating fat tails and asymmetry. This confirms that a standard normal distribution is an inadequate assumption for these returns.

### Damage

Models that assume normality — such as the standard Black-Scholes option pricing model — will systematically **underestimate tail risk**. This leads to mispricing of out-of-the-money put options (which hedge against crashes), under-reservation of economic capital, and inaccurate Value-at-Risk (VaR) estimates, all of which can expose the firm to catastrophic, unhedged losses during market dislocations.

### Directions

To address skewness in volatility modeling, practitioners should:
- Replace the normality assumption with a **Student's t-distribution** or **skewed-t distribution** (Hansen, 1994) within GARCH-family models.
- Use **EGARCH** or **GJR-GARCH** models, which explicitly model the asymmetric response of volatility to positive vs. negative shocks (the leverage effect).
- Apply **Cornish-Fisher expansion** to VaR calculations to adjust for skewness and excess kurtosis.

## Challenge 2: Sensitivity to Outliers

### Definition

An observation $x_i$ is classified as an outlier using the Interquartile Range (IQR) rule:

$$x_i \text{ is an outlier if } \quad x_i < Q_1 - 1.5 \times IQR \quad \text{or} \quad x_i > Q_3 + 1.5 \times IQR$$

where $IQR = Q_3 - Q_1$, and $Q_1$, $Q_3$ are the 25th and 75th percentiles respectively. Equivalently, a Z-score criterion flags observations where:

$$Z_i = \frac{x_i - \bar{x}}{s} > 3$$

### Description

Financial time series are prone to sudden shocks driven by macroeconomic events, earnings surprises, or market crises. Standard estimators such as the sample mean and variance are **non-robust** — a single extreme observation can disproportionately distort parameter estimates, misrepresent the true underlying return process, and undermine the reliability of any model trained on such data.

### Diagnosis

Outliers can be identified through:
1. **Visual inspection** — Boxplots and time-series plots reveal extreme data points.
2. **IQR rule** — Flags observations beyond 1.5×IQR from the quartiles.
3. **Z-score criterion** — Observations with |Z| > 3 are flagged as extreme.
4. **Cook's Distance** — In regression contexts, measures the influence of each observation on fitted values.

```python
# ── DEMONSTRATION: Sensitivity to Outliers ───────────────────────────────────

returns = data['Returns']

# IQR method
q1, q3 = returns.quantile(0.25), returns.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outlier_mask = (returns < lower_bound) | (returns > upper_bound)
outlier_count = outlier_mask.sum()

# Z-score method
z_scores = np.abs(stats.zscore(returns))
z_outlier_count = (z_scores > 3).sum()

# Impact on mean and std: with vs without outliers
mean_with    = returns.mean()
std_with     = returns.std()
mean_without = returns[~outlier_mask].mean()
std_without  = returns[~outlier_mask].std()

print("=" * 55)
print("   OUTLIER DIAGNOSTIC REPORT — AAPL Returns")
print("=" * 55)
print(f"  Total Observations          : {len(returns)}")
print(f"  IQR Lower Bound             : {lower_bound:.5f}")
print(f"  IQR Upper Bound             : {upper_bound:.5f}")
print(f"  Outliers (IQR method)       : {outlier_count} ({100*outlier_count/len(returns):.2f}%)")
print(f"  Outliers (|Z| > 3 method)   : {z_outlier_count} ({100*z_outlier_count/len(returns):.2f}%)")
print("-" * 55)
print(f"  Mean  WITH outliers         : {mean_with:.6f}")
print(f"  Mean  WITHOUT outliers      : {mean_without:.6f}")
print(f"  Std   WITH outliers         : {std_with:.6f}")
print(f"  Std   WITHOUT outliers      : {std_without:.6f}")
print(f"  Volatility inflation factor : {std_with/std_without:.4f}x")
print("=" * 55)
print("INTERPRETATION: Outliers inflate estimated volatility.")
print("Removing them reduces std, showing their outsized impact.")
```

- **Jump-diffusion models** (e.g., Merton 1976): Explicitly model outliers as a Poisson jump process superimposed on continuous Brownian motion, providing a theoretically justified treatment of extreme events.

## Challenge 3: Multicollinearity

### Definition

Multicollinearity is quantified using the **Variance Inflation Factor (VIF)** for the $i$-th predictor:

$$VIF_i = \frac{1}{1 - R_i^2}$$

where $R_i^2$ is the coefficient of determination from regressing predictor $X_i$ on all remaining predictors. The **condition number** $\kappa$ of the design matrix $\mathbf{X}$ provides an additional measure:

$$\kappa = \frac{\lambda_{\max}}{\lambda_{\min}}$$

where $\lambda_{\max}$ and $\lambda_{\min}$ are the largest and smallest eigenvalues of $\mathbf{X}^\top\mathbf{X}$. Values of $\kappa > 30$ indicate severe multicollinearity.

### Description

Multicollinearity arises when two or more predictor variables in a regression model are highly linearly correlated. In financial models, this commonly occurs when multiple technical indicators (e.g., several moving averages) or macroeconomic factors (e.g., GDP growth and industrial production) are included simultaneously — making it statistically impossible to isolate the independent contribution of each factor to the response variable.

### Diagnosis

1. **Correlation Matrix**: Pairwise correlations exceeding |0.8| suggest potential multicollinearity.
2. **VIF scores**: VIF > 5 warrants attention; VIF > 10 indicates severe multicollinearity requiring remediation.
3. **Condition Number**: Values above 30 signal ill-conditioning of the design matrix.

```python
# ── DEMONSTRATION: Multicollinearity ─────────────────────────────────────────

# Build a feature matrix with multiple moving averages — classic collinear predictors
data['MA5']  = data['Close'].rolling(5).mean()
data['MA10'] = data['Close'].rolling(10).mean()
data['MA20'] = data['Close'].rolling(20).mean()
data['MA50'] = data['Close'].rolling(50).mean()

features = data[['MA5', 'MA10', 'MA20', 'MA50']].dropna()

# Correlation matrix
corr_matrix = features.corr()

# VIF calculation
X_vif = features.assign(const=1)
vif_data = pd.DataFrame({
    'Feature': features.columns,
    'VIF'    : [variance_inflation_factor(X_vif.values, i) for i in range(len(features.columns))]
})

# Condition number
X_scaled = (features - features.mean()) / features.std()
eigenvalues = np.linalg.eigvals(X_scaled.T @ X_scaled)
condition_number = np.sqrt(eigenvalues.max() / eigenvalues.min())

print("=" * 55)
print("   MULTICOLLINEARITY DIAGNOSTIC REPORT")
print("=" * 55)
print("\nCorrelation Matrix:")
print(corr_matrix.round(4).to_string())
print("\nVariance Inflation Factors:")
print(vif_data.to_string(index=False))
print(f"\nCondition Number (κ)        : {condition_number:.2f}")
print("-" * 55)
print("RESULT: All VIF scores >> 10 and κ >> 30.")
print("Severe multicollinearity confirmed.")
print("=" * 55)
```

This makes the model uninterpretable — a trader cannot determine which timeframe of momentum is driving the signal — and unreliable for out-of-sample prediction, as coefficients may not generalize.

### Directions

- **Principal Component Analysis (PCA)**: Decompose the correlated feature matrix into orthogonal principal components that retain maximum variance. Using the top $k$ components as predictors eliminates collinearity by construction.

- **Lasso Regression (L1 regularization)**: Adds penalty $\lambda \|\boldsymbol{\beta}\|_1$, performing automatic feature selection by setting redundant coefficients to exactly zero.
- **Feature engineering**: Replace multiple moving averages with a single momentum signal or use differenced/ratio features that capture rate-of-change rather than raw levels.

## Challenge 4: Overfitting

### Definition

Overfitting occurs when a model's out-of-sample (generalization) error substantially exceeds its in-sample (training) error. Formally, using Mean Squared Error (MSE) as the loss function:

$$\text{Overfitting} \iff MSE_{\text{test}} \gg MSE_{\text{train}}$$

This can be decomposed through the **Bias-Variance Tradeoff**:

$$E\left[(y - \hat{f}(x))^2\right] = \underbrace{\text{Bias}^2[\hat{f}]}_{\text{underfitting error}} + \underbrace{\text{Var}[\hat{f}]}_{\text{overfitting error}} + \underbrace{\sigma^2_{\varepsilon}}_{\text{irreducible noise}}$$

An overfitted model has near-zero bias but very high variance: it memorizes training noise rather than learning the underlying signal.

### Description

Overfitting occurs when a model is excessively complex — with too many parameters relative to the number of observations — causing it to capture random noise in the training data rather than the true underlying financial signal. The result is a model that performs impressively in backtests but fails immediately when deployed in live markets, because historical noise does not repeat.

### Diagnosis

1. **Train/Test Split**: Compare MSE on training data vs. held-out test data. A large gap confirms overfitting.
2. **Time-Series Cross-Validation**: Use walk-forward (expanding or rolling window) cross-validation to obtain a realistic out-of-sample performance estimate while respecting temporal ordering.
3. **Learning curves**: Plot training and validation error as a function of training set size — overfitting manifests as a persistent gap between the two curves.

```python
# ── DEMONSTRATION: Overfitting ───────────────────────────────────────────────

# Use the last 120 trading days of AAPL close prices
price_series = data['Close'].values[-120:]
x_all        = np.arange(len(price_series)).reshape(-1, 1)

# Train/test split (80/20, respecting time order)
split_idx = int(len(price_series) * 0.80)
x_train, x_test = x_all[:split_idx], x_all[split_idx:]
y_train, y_test = price_series[:split_idx], price_series[split_idx:]

results = []
degrees = [1, 3, 7, 12, 15]

for deg in degrees:
    poly  = PolynomialFeatures(degree=deg)
    x_tr_p = poly.fit_transform(x_train)
    x_te_p = poly.transform(x_test)

    model = LinearRegression().fit(x_tr_p, y_train)
    mse_train = mean_squared_error(y_train, model.predict(x_tr_p))
    mse_test  = mean_squared_error(y_test,  model.predict(x_te_p))

    results.append({'Degree': deg, 'Train MSE': mse_train, 'Test MSE': mse_test,
                    'Gap (Test-Train)': mse_test - mse_train})

results_df = pd.DataFrame(results)

print("=" * 65)
print("   OVERFITTING DIAGNOSTIC REPORT — AAPL Close Price Model")
print("=" * 65)
print(results_df.to_string(index=False, float_format=lambda x: f'{x:>12.2f}'))
print("-" * 65)
print("INTERPRETATION:")
print("  As polynomial degree increases, Train MSE falls toward zero.")
print("  Test MSE initially improves but then explodes at high degrees.")
print("  The widening gap is the fingerprint of overfitting.")
print("=" * 65)
```

The more parameters a model contains without adequate regularization, the more severe this problem becomes.

### Directions

- **Time-Series Cross-Validation (Walk-Forward)**: Split data into sequential train/validation folds, ensuring the model is always tested on data it has never seen. This provides a realistic estimate of live performance.
- **Regularization — Ridge (L2) and Lasso (L1)**: Add a complexity penalty to the loss function to shrink coefficients and prevent the model from memorizing noise. The regularized objective is: $\min_{\boldsymbol{\beta}} \|\mathbf{y} - \mathbf{X}\boldsymbol{\beta}\|^2 + \lambda\|\boldsymbol{\beta}\|^p$ where $p=2$ (Ridge) or $p=1$ (Lasso).
- **Information Criteria (AIC/BIC)**: Use Akaike (AIC) or Bayesian (BIC) information criteria to select model order — these criteria penalize model complexity, naturally trading off goodness-of-fit against parsimony.
- **Parsimony principle**: Prefer simpler models (e.g., ARMA(1,1)-GARCH(1,1)) over high-dimensional alternatives unless additional complexity is strongly justified by out-of-sample evidence.

## Step 4: Synthesis — The Central Challenge: Skewness

### Identification of the Core Challenge

Among the four challenges addressed in this handbook, **skewness** is identified as the most foundational problem for derivatives desk practitioners. While all four challenges are serious, skewness is the root cause that cascades into the others: non-normal, asymmetric return distributions amplify the damage from outliers, invalidate the normality assumptions underpinning multicollinearity tests, and make it harder to evaluate model generalization (overfitting detection) because standard MSE metrics weight all errors equally regardless of directional asymmetry.

### Why Skewness is the Priority Challenge

The Black-Scholes model — the foundation of modern derivatives pricing — explicitly assumes that log returns follow a **symmetric normal distribution** with constant volatility. Our empirical analysis of AAPL returns (2018–2025) conclusively rejects this assumption: the Jarque-Bera test produces a p-value effectively equal to zero, and the computed skewness is significantly negative. This is not a quirk of AAPL specifically; negative skewness in equity returns is a well-documented stylized fact of financial markets (Cont, 2001).

The practical consequences are severe:

1. **Option mispricing**: Black-Scholes underprices out-of-the-money puts and overprices out-of-the-money calls when returns are negatively skewed. This creates the empirically observed **implied volatility smirk** — a phenomenon that the desk must account for in any production pricing model (Hull, 2021).

2. **VaR underestimation**: Value-at-Risk computed under normality systematically underestimates the probability of large losses. During the 2008 financial crisis and the March 2020 COVID crash, VaR models built on normal distributions failed catastrophically across the industry.

3. **Inadequate hedging**: Delta and gamma hedging ratios derived from Black-Scholes are incorrect when the true return distribution is skewed, leaving residual directional risk that is unaccounted for.

### Recommended Resolution

Building directly on the Directions section for skewness above, the recommended model upgrade path for the derivatives desk is:

**Short-term (immediate)**: Implement a **GJR-GARCH(1,1)** model with a **skewed Student's t-distribution** for innovations. This captures three key empirical features simultaneously: volatility clustering (GARCH), the leverage effect — negative shocks increase volatility more than positive shocks of equal magnitude (GJR asymmetric term), and fat tails with negative skewness (skewed-t distribution). The GJR-GARCH variance equation is:

$$\sigma_t^2 = \omega + \alpha \varepsilon_{t-1}^2 + \gamma \varepsilon_{t-1}^2 \mathbf{1}_{[\varepsilon_{t-1}<0]} + \beta \sigma_{t-1}^2$$

where $\gamma > 0$ captures the leverage effect. This model is well-supported in the academic literature as a significant improvement over symmetric GARCH for equity returns (Glosten, Jagannathan, and Runkle, 1993).

**Medium-term**: Construct a **stochastic volatility model with jumps** (SVJ), which models skewness as arising from two distinct sources: the leverage correlation between returns and volatility innovations, and a Poisson jump component capturing sudden shocks. This provides a theoretically richer framework for pricing exotic derivatives.

**For VaR/risk management**: Apply the **Cornish-Fisher expansion** to adjust quantile estimates for observed skewness ($S$) and excess kurtosis ($K$):

$$z_{CF} = z + \frac{z^2-1}{6}S + \frac{z^3-3z}{24}K - \frac{2z^3-5z}{36}S^2$$

This provides a first-order correction to normal-distribution VaR without requiring a full distributional assumption change.



# Explicit Mapping of Directions to Alleviate Damage

---

## 1. Multicollinearity

**Damage Recap:** Inflated standard errors destabilize hedging ratios, causing traders to over-hedge or under-hedge. Even minor data changes can produce wild swings in model outputs.

**Direction:** The core fix is removing redundant information from inputs. Two practical approaches work well here:

- **PCA (Principal Component Analysis):** Transforms correlated inputs into orthogonal (independent) components, eliminating overlap between predictors.
- **VIF-based feature pruning:** Iteratively drop features whose Variance Inflation Factor exceeds a set threshold, keeping only variables that contribute unique information.

Both approaches directly stabilize coefficient estimates, so small shifts in market data no longer cascade into erratic changes in hedging ratios.

---

## 2. Sensitivity to Outliers

**Damage Recap:** Standard OLS models over-accommodate rare shocks, inflating tail-risk estimates and pushing the desk into excessive capital allocation during normal market conditions — quietly eroding profitability.

**Direction:** Replace the quadratic penalty of OLS with methods that cap outlier influence:

- **Huber Regression / LAD (Least Absolute Deviations):** Apply linear rather than squared penalties to large residuals, so extreme observations contribute less leverage over model parameters.
- **Winsorization (1st–99th percentile):** Clips extreme values to boundary thresholds, preserving the *direction* of signals without letting them distort model coefficients.

Either approach restores realistic tail-risk estimates and rebuilds profitability during calm trading conditions.

---

## 3. Skewness

**Damage Recap:** Assuming symmetric return distributions causes systematic underpricing of out-of-the-money put options and leaves the desk exposed to severe drawdowns during market stress.

**Direction:** Move away from the symmetric normal assumption by explicitly modelling asymmetry:

- **GARCH with skewed Student's t errors:** Captures both volatility clustering and the heavier, asymmetric left tail of equity returns.
- **SABR / Stochastic Volatility models:** Embed skew directly into the options pricing framework, ensuring the volatility smile is correctly reproduced.

Properly weighting the left tail means crash probabilities are no longer underestimated, and downside risk is priced accurately.

---

## 4. Overfitting

**Damage Recap:** Models that memorize historical noise show artificially low backtest errors but collapse in live trading — producing mispriced exotic derivatives and large hedging slippages.

**Direction:** Two complementary controls enforce genuine generalization:

- **Walk-forward (out-of-sample) cross-validation:** The model is always tested on data it has never seen during training, giving an honest picture of live-market performance.
- **Lasso (L₁) regularization:** Penalizes model complexity by shrinking irrelevant coefficients to exactly zero, forcing the model to ignore historical noise rather than fit it.

Together, these controls ensure the model reflects real market structure, not the idiosyncrasies of the training window.

---

## Step 4 — Identifying the Solved Challenge: *Prioritizing Factors*

In quantitative finance, quants routinely face hundreds of candidate predictors — macro variables, technical indicators, sentiment signals — and must determine which genuinely drive the asset versus which are noise.

This challenge of **factor prioritization** is comprehensively addressed as a by-product of the Multicollinearity and Overfitting work above:

| Protocol | Mechanism | Effect on Factor Prioritization |
|---|---|---|
| **VIF analysis / PCA** | Eliminates redundant variables | Retains only features carrying unique market information |
| **Lasso regularization** | Shrinks irrelevant coefficients to zero | Acts as an automated feature-selection engine |

By combining **redundancy elimination** (Multicollinearity protocols) with **automated shrinkage** (Overfitting protocols), the desk obtains a mathematically rigorous, reproducible framework for deciding which factors deserve a place in the model — and which should be discarded.


### Conclusion

The empirical analysis in this handbook demonstrates that standard normal-distribution assumptions are untenable for AAPL equity returns. All four challenges — skewness, outlier sensitivity, multicollinearity, and overfitting — must be actively managed in any production volatility model. However, skewness represents the foundational challenge: correctly modeling the asymmetric shape of the return distribution is a prerequisite for reliable option pricing, risk measurement, and hedging. The GJR-GARCH model with skewed-t innovations provides a practically implementable, theoretically grounded solution that the derivatives desk can adopt immediately.
